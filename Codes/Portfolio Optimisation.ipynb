```python
combi1 = combine.loc[:,["xrp","usdt","usdc","voo","vea","vgt"]] #half crypto, half etfs
combi2 = combine.loc[:,["luna","eth","btc","bsv","mbb","aoa"]] #half crypto, half etfs
combi3 = combine.loc[:,["luna","ada","btc","bsv","sh","tqqq"]] #half crypto half etfs
combi4 = combine.loc[:,["luna","bnb","btc","tqqq","mbb","voo"]] #half cryto half etfs 
combi8 = combine.loc[:,["ada","bnb","eth","tqqq","vgt","sh"]] #half cryto half etfs 

bsv_combi = combine.loc[:,["bsv","bnb","btc","eth","xrp","ada"]] #short term + crypto
mbb_combi = combine.loc[:,["mbb","bnb","btc","eth","xrp","ada"]] #long term + crypto
voo_combi = combine.loc[:,["luna","bnb","btc","ada","voo"]] #us equity etf

only_etfs = combine.loc[:,["voo", "vea","vgt","bsv","mbb","aoa","tqqq", "sh"]]
```

```python
combi1_log_returns = np.log(combi1/combi1.shift())

combi2_log_returns = np.log(combi2/combi2.shift())

combi3_log_returns = np.log(combi3/combi3.shift())

combi4_log_returns = np.log(combi4/combi4.shift())

combi8_log_returns = np.log(combi8/combi8.shift())

combine_log_returns  = np.log(combine/combine.shift())

bsv_combi  = np.log(bsv_combi/bsv_combi.shift()) #short term + crypto
 
mbb_combi = np.log(mbb_combi/mbb_combi.shift()) #long term + crypto

voo_combi = np.log(voo_combi/voo_combi.shift())#us equity etf + crypto

only_etfs_log = np.log(only_etfs/only_etfs.shift())

only_crypto_log = np.log(crypto/crypto.shift())
```

```python
def getcombination():
    import random

    cryptolist = ["xrp","usdt","usdc","ada","eth","luna","btc","bnb"]
    etflist = ["aoa","vgt","vea","mbb","tqqq","sh","bsv","voo"]

    combinationlist = []

    number_of_funds = random.randint(4,10)

    for i in range(number_of_funds):
        index = random.randint(0,7)
        whichlist = random.randint(1,2)
        if whichlist == 1:
            while cryptolist[index] in combinationlist:
                index = random.randint(0,7)
            combinationlist.append(cryptolist[index])
        else:
            while etflist[index] in combinationlist:
                index = random.randint(0,7)
            combinationlist.append(etflist[index])

    combinationlist = combine.loc[:,combinationlist]
    combinationlist = np.log(combinationlist/combinationlist.shift())
    return (combinationlist)

def simulation(combi):
    n = 5000
    
    weights = np.zeros((n, len(combi.columns)))
    exp_rtns = np.zeros(n)
    exp_vols = np.zeros(n)
    sharpe_ratios = np.zeros(n)

    for i in range(n):
        weight= np.random.random(len(combi.columns))
        weight /= weight.sum()
        weights[i] = weight

        exp_rtns[i] = np.sum(combi.mean()*weight)*252
        exp_vols[i] = np.sqrt(np.dot(weight.T, np.dot(combi.cov()*252, weight)))
        sharpe_ratios[i] = exp_rtns[i] / exp_vols[i]
    
    %matplotlib notebook
 
    fig, ax = plt.subplots()
    ax.scatter(exp_vols, exp_rtns, c=sharpe_ratios)
    ax.scatter(exp_vols[sharpe_ratios.argmax()], exp_rtns[sharpe_ratios.argmax()], c='r')
    ax.set_xlabel('Expected Volatility')
    ax.set_ylabel('Expected Return')
    print(combi.columns)
    print("weights", weights[sharpe_ratios.argmax()] * 100)
    print("sharpe ratio max",sharpe_ratios.max())
    print("expected volatility and expected returns :" ,exp_vols[sharpe_ratios.argmax()] , exp_rtns[sharpe_ratios.argmax()] * 100)


x = getcombination()
simulation(x)
```
